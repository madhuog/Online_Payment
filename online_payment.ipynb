```python
import datetime                                          # to extract the  present date from the system
import random                                           # to generate random numbers for OTP

print("*********Wel_come to online_payment*********")
print("Only payment through card is accepted")


# to extract present date from the system and convert it to string

date=datetime.date.today()         # present date
year=str(date.year)               #extacting year and converting it into string
month=str(date.month)             #extracting month and converting it into string
if len(month)<2:
    month=month.zfill(2)          #adding 'zero' before single digit month
    
    
# Card_number_details 

f=0                                                      
while f==0:
    card_num=input("\nEnter your 16 digit card number:")
    if len(card_num)!=16 or card_num.isnumeric()==False:
        print("Invalid, Please check with card number")
    else:
        f=1
        

        
#Card_validity date details :

g=0 
while g==0:
    exp_year_str=str(input("\nEnter the validity year(YYYY):"))
    if len(exp_year_str)!=4 or exp_year_str<year or exp_year_str.isnumeric()==False:
        print("invalid format(YYYY) or card expired ")
    else:
        #print(exp_year)
        g=1
h=0
while h==0:
    exp_month=str(input("\nEnter the validity month(MM):"))
    #exp_month_str=str(exp_month)
    if exp_year_str==year and (len(exp_month)!=2  or (exp_month<month or exp_month>=str(13)) or exp_month.isnumeric()==False):
        print("invalid format(MM) or card expired")
    else:
        h=1

        
        
# Card_CVV_details :


i=0
while i==0:
    cvv=str(input("\nEnter the 3 digit CVV:"))
    if len(cvv)!=3 or cvv.isnumeric()==False:
        print("Enter valid cvv")
    else:
        i=1
        
print("\nCARD DETAILS:")        
print("Card_Number:",card_num)
print("Exp_date(MM/YYYY):{}/{}".format(exp_year_str,exp_month))
print("CVV:",cvv)

choice=input("\nWould you like to make payment(Yes/No):")
choice=choice.upper()
j=0
k=0
if choice=="YES" or choice=='Y':
    rs=input('Enter the amount to be paid:')
    otp=random.randint(1000,9999)                      #OTP generation
    print("\n\n{} is your OTP to proceed with transaction".format(otp))
    while j==0 and k<3:
        entered_otp=int(input("\nPlease enter the OTP you recived to your registerd phone number:"))
        if entered_otp!=otp:
            print("invalid OTP,please enter the correct OTP(max 3 chance)")
            k+=1
            if k>2:
                print("Maximum attempted reached")
        else:
            j=1
            print('\nYour Online Payment of Rs.',rs,'/- is Successful.Thank You.')
            
        
else:
    print('Payment failed')
    print("\nThank You")
    
    
```
